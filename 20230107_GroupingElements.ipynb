```python
def root_grouping(networks):

    count = 1
    clusters = {count:[networks[0]['Endpoint_A'], networks[0]['Endpoint_B']]}
    n = len(networks)
    original_size = len(networks)

    while(n>0):
        for network in networks:
            var_1 = clusters[count] + [network['Endpoint_A'], network['Endpoint_B']]
            var_2 = list(set(var_1))

            if(len(var_1) > len(var_2)):
                clusters[count] = var_2
                networks.remove(network)

        if(len(networks)==n):
            count += 1
            clusters[count] = [networks[0]['Endpoint_A'], networks[0]['Endpoint_B']]

        n = len(networks)
        percent_completed = 1 - n/original_size
        print(percent_completed)

    return clusters
```

```python
def create_network():
    alph = string.ascii_uppercase

    network = {
        'Endpoint_A':alph[randrange(26)]+str(randrange(5000)),
        'Endpoint_B':alph[randrange(26)]+str(randrange(5000))
    }

    return network
```

```python
sample_network_2 = []

for k in range(100000):
    sample_network_2.append(create_network())
```
